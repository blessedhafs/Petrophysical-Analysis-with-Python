```python
fig, axes = plt.subplots(figsize= (10, 10))
curve_names = ["Gamma", "Resistivity", "Density"]

# set up the plot axes
ax1 = plt.subplot2grid((1, 3), (0, 0), rowspan = 1, colspan = 1)
ax2 = plt.subplot2grid((1, 3), (0, 1), rowspan = 1, colspan = 1)
ax3 = plt.subplot2grid((1, 3), (0, 2), rowspan = 1, colspan = 1)

# set up individual subplots/track
ax1.plot("GR", "DEPTH", data = df, color = "green")
ax1.set_xlim(0, 200)

ax2.plot("RES", "DEPTH", data =df, color = "red")
ax2.set_xlim(0.2, 2000)
ax2.semilogx()

ax3.plot("DEN", "DEPTH", data = df, color = "purple")
ax3.set_xlim(1.9, 2.5)

# set up the common element between the subplots
for i, ax in enumerate(fig.axes):
    ax.set_ylim(7500, 6850) # sets the depth range
    ax.grid()
    ax.set_xlabel(curve_names[i])

# Hide tick labels on the y-axis
for ax in [ax2, ax3]:
    plt.setp(ax.get_yticklabels(), visible = False)

# reduce the space between each subplot
fig.subplots_adjust(wspace = 0.05)
```

```python
fig, axes = plt.subplots(figsize= (10, 10))
curve_names = ["Gamma", "Resistivity", "Density","Neutron"]

# set up the plot axes
ax1 = plt.subplot2grid((1, 3), (0, 0), rowspan = 1, colspan = 1)
ax2 = plt.subplot2grid((1, 3), (0, 1), rowspan = 1, colspan = 1)
ax3 = plt.subplot2grid((1, 3), (0, 2), rowspan = 1, colspan = 1)
ax4 = ax3.twiny()

# set up the plot individual subplots/track
ax1.plot("GR", "DEPTH", data = df, color = "green")
ax1.set_xlim(0, 200)

ax2.plot("RES", "DEPTH", data =df, color = "red")
ax2.set_xlim(0.2, 2000)
ax2.semilogx()

ax3.plot("DEN", "DEPTH", data = df, color = "purple")
ax3.set_xlim(1.9, 2.5)

ax4.plot("NEU", "DEPTH", data = df, color = "blue")
ax4.set_xlim(0.5, 0)

# set up the common element between the subplots
for i, ax in enumerate(fig.axes):
    ax.set_ylim(7500, 6850) # sets the depth range
    
    ax.xaxis.set_ticks_position("top")
    ax.xaxis.set_label_position("top")
    ax.set_xlabel(curve_names[i])
    
    if i == 3:
        ax.spines["top"].set_position(("axes", 1.08))
    else:
        ax.grid()

# Hide tick labels on the y-axis
for ax in [ax2, ax3]:
    plt.setp(ax.get_yticklabels(), visible = False)

# reduce the space between each subplot
fig.subplots_adjust(wspace = 0.05)
```

```python
fig, axes = plt.subplots(figsize= (10, 10))
curve_names = ["Gamma", "Resistivity", "Density","Neutron"]

# set up the plot axes
ax1 = plt.subplot2grid((1, 3), (0, 0), rowspan = 1, colspan = 1)
ax2 = plt.subplot2grid((1, 3), (0, 1), rowspan = 1, colspan = 1)
ax3 = plt.subplot2grid((1, 3), (0, 2), rowspan = 1, colspan = 1)
ax4 = ax3.twiny()

# set up the plot individual subplots/track
ax1.plot("GR", "DEPTH", data = df, color = "green", lw = 0.5) #reduce the thickness of the lines
ax1.set_xlim(0, 200)

ax2.plot("RES", "DEPTH", data =df, color = "red", lw = 0.5)
ax2.set_xlim(0.2, 2000)
ax2.semilogx()

ax3.plot("DEN", "DEPTH", data = df, color = "purple", lw = 0.5)
ax3.set_xlim(1.9, 2.5)

ax4.plot("NEU", "DEPTH", data = df, color = "blue", lw = 0.5)
ax4.set_xlim(0.5, 0)

# set up the common element between the subplots
for i, ax in enumerate(fig.axes):
    ax.set_ylim(7500, 6850) # sets the depth range
    
    ax.xaxis.set_ticks_position("top")
    ax.xaxis.set_label_position("top")
    ax.set_xlabel(curve_names[i])
    
    if i == 3:
        ax.spines["top"].set_position(("axes", 1.08))
    else:
        ax.grid()
        
# Hide tick labels on the y-axis
for ax in [ax2, ax3]:
    plt.setp(ax.get_yticklabels(), visible = False)

# reduce the space between each subplot
fig.subplots_adjust(wspace = 0.05)
```
