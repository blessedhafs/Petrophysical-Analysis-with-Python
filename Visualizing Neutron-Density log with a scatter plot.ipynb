```python
plt.scatter(x = "NEU", y = "DEN", data =df)

# sets a scale limit for each axis
plt.xlim(0, 0.45) 
plt.ylim(3, 1.5) # flips density axis in reverse order

plt.xlabel("Neutron Porosity (NEU) - %", fontsize = 14)
plt.ylabel("Bulk Density (DEN) - g/cc", fontsize = 14)

plt.show()
```

```python
# add color using the GR and set range from 0 to 100
plt.scatter(x = "NEU", y = "DEN", data =df, c = "GR", vmin = 0, vmax = 100)

# sets a scale limit for each axis
plt.xlim(0, 0.45) 
plt.ylim(3, 1.5) # flips density axis in reverse order

plt.xlabel("Neutron Porosity (NEU) - %", fontsize = 14)
plt.ylabel("Bulk Density (DEN) - g/cc", fontsize = 14)

plt.show()
```

```python
# list of available colormaps
from matplotlib import colormaps
list(colormaps)
```

```python
# add color using the GR and set range from 0 to 100
# pass cmap for interpretability of the colors
plt.scatter(x = "NEU", y = "DEN", data =df, c = "GR", vmin = 0, vmax = 100, cmap= "rainbow")

# sets a scale limit for each axis
plt.xlim(0, 0.45) 
plt.ylim(3, 1.5) # flips density axis in reverse order

plt.xlabel("Neutron Porosity (NEU) - %", fontsize = 14)
plt.ylabel("Bulk Density (DEN) - g/cc", fontsize = 14)

# add a colorbar to understand each color
plt.colorbar(label = "Gamma Ray - API")

plt.show()
```
